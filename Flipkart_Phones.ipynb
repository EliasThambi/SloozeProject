## Reading the Data


```
from pyspark.sql.types import * 
from pyspark.sql.functions import * 
df = spark.read.format('csv').\
option('inferSchema',True).\
option('header',True).\
load('/FileStore/rawdata/FlipKartPhones.csv')
```

```
df.display()
```

## Removing unwanted columns

```
# Filtering a particular row because of bad data
df_new = df.drop("web-scraper-start-url", "links-href","pagelinks-href").\
        withColumnRenamed("web-scraper-order","Id").\
        filter(col("Id") != '1749482734-148')
df_new.display()
```

### Count of all Different Processors

```
 brand_count = df_new.groupBy("ProcessBrand").count()
 brand_count.display()
```

## Most popular phones Table

```
max_rated_phn = df_new.filter(col("Rating")>4.5)
max_rated_phn.display()
```

## Most popular phones 

```
max_rated_phn = df_new.filter(col("Rating")>4.5)
max_rated_phn.display()
```

## Phones with least cost and max discount

```
df_new.sort(['ActualCost','DiscountPercentage'], ascending = [1,0]).display()
```

## Count of phones according to storage table

```
# Type casting internal storage after spliting it

df_temp = df_new.withColumn('InternalStorage',split('InternalStorage',' ')[0]).\
    withColumn('InternalStorage', col('InternalStorage').cast(IntegerType()))

df_temp.groupBy("InternalStorage", "RAM").agg(count("*").alias("PhoneCount")).\
    sort(["InternalStorage"],ascending = [0]).\
    display()
```

## Count of phones according to storage Line Graph

```
df_temp = df_new.withColumn('InternalStorage',split('InternalStorage',' ')[0]).\
    withColumn('InternalStorage', col('InternalStorage').cast(IntegerType()))

df_temp.groupBy("InternalStorage", "RAM").agg(count("*").alias("PhoneCount")).\
    sort(["InternalStorage"],ascending = [0]).\
    display()
```
